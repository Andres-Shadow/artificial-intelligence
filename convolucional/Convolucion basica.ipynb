```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('./20.png')
```

```python
img_array = np.asarray(img)
#img_array = np.array(img) / 255.0  # Normalizar valores entre 0 y 1
```

```python
blur_filter = np.array([[-1, 0 , 1],
                        [-2, 0 , 2],
                        [-1, 0 , 1]])
```

```python
from scipy.signal import convolve2d
# Separar los canales de color
r, g, b = img_array[:, :, 0], img_array[:, :, 1], img_array[:, :, 2]

# Aplicar la convoluci√≥n a cada canal de color
blurred_r = convolve2d(r, blur_filter, mode='same', boundary='symm')
blurred_g = convolve2d(g, blur_filter, mode='same', boundary='symm')
blurred_b = convolve2d(b, blur_filter, mode='same', boundary='symm')
```

```python
blurred_img = np.dstack([blurred_r, blurred_g, blurred_b])

# Visualizar la imagen resultante
plt.imshow(blurred_img.astype(np.uint8))
plt.show()

# Guardar la imagen resultante
result = Image.fromarray(blurred_img.astype(np.uint8))
result.save('blurred_image.png')
```
